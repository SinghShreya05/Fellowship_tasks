```python
def rate(temp):
    lst1=[]
    add=0
    r=0
    temp=list(temp)
    #print(temp)
    #print(range(len(temp)))
    for i in range(len(temp)-1):
        res=temp[i+1]-temp[i]
        lst1.append(res)
    for x in lst1:
        add=add+x
        r=add/len(temp)
        r=r/100
    return r
```

```python
def forecasting_covid_patients(t,P_o,cases):
    r=rate(cases)
    print('time*rate is= ',t*r)
    mul=np.exp(t*r)
    print('exponential multiplication is = ',np.exp(t*r))
    P_t=P_o*mul
    val=int(P_t)
    return val 
```
